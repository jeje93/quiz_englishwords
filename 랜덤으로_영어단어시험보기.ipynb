```python
import csv
data = csv.reader(open('3word.csv'))
next(data)

cha_1 = []
cha_2 = []
cha_3 = []
cha_4 = []
cha_5 = []
cha_6 = []


for row in data:
  if '1' in row[0]:
    for a in range(1,8):
      cha_1.append(row[a])
data = csv.reader(open('3word.csv'))
for row in data:
  if '2' in row[0]:
    for a in range(1,8):
      cha_2.append(row[a])
data = csv.reader(open('3word.csv'))
for row in data:
  if '3' in row[0]:
    for a in range(1,8):
      cha_3.append(row[a])
data = csv.reader(open('3word.csv'))
for row in data:
  if '4' in row[0]:
    for a in range(1,8):
      cha_4.append(row[a])
data = csv.reader(open('3word.csv'))
for row in data:
  if '5' in row[0]:
    for a in range(1,8):
      cha_5.append(row[a])
data = csv.reader(open('3word.csv'))
for row in data:
  if '6' in row[0]:
    for a in range(1,8):
      cha_6.append(row[a])

data_ans = csv.reader(open('3word_ans.csv'))
next(data_ans)

ans_1 = []
ans_2 = []
ans_3 = []
ans_4 = []
ans_5 = []
ans_6 = []

for row in data_ans:
  if '1' in row[0]:
    for a in range(1,8):
      ans_1.append(row[a])
data_ans = csv.reader(open('3word_ans.csv'))
for row in data_ans:
  if '2' in row[0]:
    for a in range(1,8):
      ans_2.append(row[a])
data_ans = csv.reader(open('3word_ans.csv'))
for row in data_ans:
  if '3' in row[0]:
    for a in range(1,8):
      ans_3.append(row[a])
data_ans = csv.reader(open('3word_ans.csv'))
for row in data_ans:
  if '4' in row[0]:
    for a in range(1,8):
      ans_4.append(row[a])
data_ans = csv.reader(open('3word_ans.csv'))
for row in data_ans:
  if '5' in row[0]:
    for a in range(1,8):
      ans_5.append(row[a])
data_ans = csv.reader(open('3word_ans.csv'))
for row in data_ans:
  if '6' in row[0]:
    for a in range(1,8):
      ans_6.append(row[a])
    
dict1 = dict(zip(cha_1,ans_1))
dict2 = dict(zip(cha_2,ans_2))
dict3 = dict(zip(cha_3,ans_3))
dict4 = dict(zip(cha_4,ans_4))
dict5 = dict(zip(cha_5,ans_5))
dict6 = dict(zip(cha_6,ans_6))

import random
import time
print('안녕하세요~ 영어 단어 시험을 보는 게임입니다.')
number = int(input('몇단원의 단어시험을 보기 원하십니까(1~6): '))
print()
print('-------------------------------------------------')
print()
if number == 1:
  print('1단원 시험을 보겠습니다. 힘내세요^^')
  keys = list(dict1.keys())
  random.shuffle(keys)
  cnt = 0
  for english in keys:
    korean = dict1[english]
    guess = input('{}의 영어 단어의 뜻은 무엇일까요? :'.format(english))
    time.sleep(0.5)
    if guess == korean:
      print('영어 단어의 뜻이 맞습니다.')
      cnt += 1
    else:
      print('영어 단어의 뜻이 틀립니다.')
      guess = input('{}의 영어 단어의 뜻을 다시 입력해주세요 :'.format(english))

if number == 2:
  print('2단원 시험을 보겠습니다. 힘내세요^^')
  keys = list(dict2.keys())
  random.shuffle(keys)
  cnt = 0
  for english in keys:
    korean = dict2[english]
    guess = input('{}의 영어 단어의 뜻은 무엇일까요? :'.format(english))
    time.sleep(0.5)
    if guess == korean:
      print('영어 단어의 뜻이 맞습니다.')
      cnt += 1
    else:
      print('영어 단어의 뜻이 틀립니다.')
      guess = input('{}의 영어 단어의 뜻을 다시 입력해주세요 :'.format(english))

if number == 3:
  print('3단원 시험을 보겠습니다. 힘내세요^^')
  keys = list(dict3.keys())
  random.shuffle(keys)
  cnt = 0
  for english in keys:
    korean = dict3[english]
    guess = input('{}의 영어 단어의 뜻은 무엇일까요? :'.format(english))
    time.sleep(0.5)
    if guess == korean:
      print('영어 단어의 뜻이 맞습니다.')
      cnt += 1
    else:
      print('영어 단어의 뜻이 틀립니다.')
      guess = input('{}의 영어 단어의 뜻을 다시 입력해주세요 :'.format(english))

if number == 4:
  print('4단원 시험을 보겠습니다. 힘내세요^^')
  keys = list(dict4.keys())
  random.shuffle(keys)
  cnt = 0
  for english in keys:
    korean = dict4[english]
    guess = input('{}의 영어 단어의 뜻은 무엇일까요? :'.format(english))
    time.sleep(0.51)
    if guess == korean:
      print('영어 단어의 뜻이 맞습니다.')
      cnt += 1
    else:
      print('영어 단어의 뜻이 틀립니다.')
      guess = input('{}의 영어 단어의 뜻을 다시 입력해주세요 :'.format(english))

if number == 5:
  print('5단원 시험을 보겠습니다. 힘내세요^^')
  keys = list(dict5.keys())
  random.shuffle(keys)
  cnt = 0
  for english in keys:
    korean = dict5[english]
    guess = input('{}의 영어 단어의 뜻은 무엇일까요? :'.format(english))
    time.sleep(0.5)
    if guess == korean:
      print('영어 단어의 뜻이 맞습니다.')
      cnt += 1
    else:
      print('영어 단어의 뜻이 틀립니다.')
      guess = input('{}의 영어 단어의 뜻을 다시 입력해주세요 :'.format(english))

if number == 6:
  print('6단원 시험을 보겠습니다. 힘내세요^^')
  keys = list(dict6.keys())
  random.shuffle(keys)
  cnt = 0
  for english in keys:
    korean = dict6[english]
    guess = input('{}의 영어 단어의 뜻은 무엇일까요? :'.format(english))
    time.sleep(0.5)
    if guess == korean:
      print('영어 단어의 뜻이 맞습니다.')
      cnt += 1
    else:
      print('영어 단어의 뜻이 틀립니다.')
      guess = input('{}의 영어 단어의 뜻을 다시 입력해주세요 :'.format(english))

print()
time.sleep(2)
score = round(cnt/len(dict1) * 100, 1)
print('-------------------------------------------------')
print('100점 만점 {} 점입니다.'.format(score))

```
